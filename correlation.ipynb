```python
#stats.kendalltau(ranking_table_all_in['NDR'], ranking_table_all_in['TR'])
import numpy as np

def get_corrs(df):
    col_correlations = df.corr(method='kendall')
    col_correlations.loc[:, :] = np.tril(col_correlations, k=-1)
    cor_pairs = col_correlations.stack()
    return cor_pairs

my_corrs = get_corrs(ranking_table_all_in)
# and the following line to retrieve the single correlation
my_corrs
```
