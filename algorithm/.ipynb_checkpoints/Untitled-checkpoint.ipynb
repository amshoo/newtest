```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
cent = pd.read_csv('evaluation/rank_result/kc--/cent/cent_kc_network.csv')
ds = pd.read_csv('evaluation/rank_result/kc--/ds/ds_kc.csv')
cent = pd.concat([ds,cent], axis=1, sort=False)
cent = cent.drop(['Unnamed: 0'], axis=1)
cent['TR_i']= cent['TR_i']*0.1
cent['DS_i']= cent['DS_i']+0.1

# Create a plots for density.
def density_plot():
    result = cent.sort_values('BC')
    result1 = cent.sort_values('CNC')
    result2 = cent.sort_values('PR')    
    result3 = cent.sort_values('TR')
    result4 = cent.sort_values('DS')
    fig, (ax1) = plt.subplots(1, sharex=True)
    ax1.set(title='Network Density', ylabel='TR_i')
    ax1.grid()
    ax1.plot(result.BC, result.BC_i, '>-')
    ax1.plot(result3.TR, result3.TR_i, '1-')
    ax1.plot(result1.CNC, result1.CNC_i, '+-')
    ax1.plot(result1.DS, result1.DS_i, '--')
    ax1.set(xlabel='Nodes', ylabel='Density')
    #save = plt.savefig("ds/---.png")
    plot = plt.show()
```

```python
density_plot()
```
