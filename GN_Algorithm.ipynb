```python
import networkx as nx
import operator
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
def graph():    
    import graph
    Graph = graph.H
    return Graph
G= graph()
#G=nx.karate_club_graph()
```

```python
sorted_bc=[1]
j=[]
actual_num_components=1
while not sorted_bc==[]:
    d_edges=nx.edge_betweenness_centrality(G)
    sorted_bc = sorted(d_edges.items(), key=operator.itemgetter(1))
    e=sorted_bc.pop()
    h=list(e[0])    
    print ("edge:", e[0],
    G.remove_edge(*e[0]))
    num_comp = nx.number_connected_components(G)
    k = h, num_comp
    j.append(k) 
    print ("R.components", num_comp)
    
    if num_comp>actual_num_components:
           actual_num_components=num_comp
```

```python
TPN = pd.DataFrame(j, columns = ('edges','l_comp'))
TPN
```

```python
plt.plot(TPN.l_comp, 'g-+')
#plt.plot(eff_table.Density, 'g-o')
plt.legend(["Efficiency"])
plt.xlabel("nodes")
plt.ylabel("Efficiency")
plt.savefig("result/email_eff1.png")
plt.show()
```
